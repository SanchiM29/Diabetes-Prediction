```python
#Importing the Dependencies
import numpy as np #numpy arrays
import pandas as pd #pandas for dataframes 
#for standarizing data to a common range (StandardScaler)
from sklearn.preprocessing import StandardScaler
#we will use train_test_split function for splitting training and testing data
from sklearn.model_selection import train_test_split
#import the model used
from sklearn import svm 
from sklearn.metrics import accuracy_score
```

```python
#Data Collection and Analysis 
#PIMA Diabetes Dataset (females only)
```

```python
#Loading the dataset using pandas to a dataframe
diabetes_dataset = pd.read_csv("diabetes.csv")
#pd.read_csv does read a csv file in dataframe
```

```python
#first five rows of the dataset
diabetes_dataset.head()
```

```python
#find out the number of rows and columns for the dataset
diabetes_dataset.shape
```

```python
#getting the statistical measures of the data
diabetes_dataset.describe()
```

```python
#To check outcome attribute, how many 0's and 1's
diabetes_dataset['Outcome'].value_counts()
```

```python
diabetes_dataset.groupby('Outcome').mean()
```

```python
#lets seperate outcome and other 8 attributes which are used for analysis ie. (seperating the labels and data)
X = diabetes_dataset.drop(columns = "Outcome", axis = 1) #dropping a column therefore axis = 1 else 0
Y = diabetes_dataset["Outcome"]
```

```python
print(X)
```

```python
print(Y)
```

```python
#Data Standarization 
scaler = StandardScaler() #taking 1 instance of the standar scaler function
```

```python
#fit the data into the scaler
scaler.fit(X)
```

```python
standarized_data = scaler.transform(X)
```

```python
print(standarized_data)
```

```python
X = standarized_data #Y remaines the same outcome 
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify = Y, random_state = 2)
#trained on X_train and then test on X_test
#0.2 represent 20% test data therefore 80% for training 
#Stratify on basis of Y proportion (0,1)
#random state is the randomness
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
#Train the model
```

```python
classifier = svm.SVC(kernel = 'linear')
```

```python
#fit training data to classifier and
#training the svm classifier
classifier.fit(X_train, Y_train) #represent the training data and their labels X and Y respectively
```

```python
#evaluate our model, how many times it is predicitng correct
#1. Accuracy on training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train) #comparing prediction with ytrain
```

```python
print("Accuracy score of the training data" , training_data_accuracy)
```

```python
#Accuracy on test data (unknown data)
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test) #comparing prediction with ytrain
```

```python
print("Accuracy score of the test data" , test_data_accuracy)
```

```python
#overfitting is when we have trained the model on training data a lot and thus we get low accuracy on test data
#therefore less accuracy on test data is not good as it is the unknown data for our prediction model
```
